```python
# 1️⃣ Kütüphaneler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2️⃣ Veri Yükleme
df = pd.read_csv("monthly_trends_2025.csv")
df.head()
```

```python
# 3️⃣ Aylara Göre Video Sayısı
video_counts = df.groupby("year_month")["n_videos"].sum().reset_index()
video_counts = video_counts.sort_values("year_month")
print(video_counts)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=video_counts, x="year_month", y="n_videos")
plt.title("Aylara Göre Video Sayısı (Toplam)")
plt.xlabel("Ay")
plt.ylabel("Video Sayısı")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 4️⃣ Aylık Ortalama Etkileşim
df["total_engagement"] = df["views"]
avg_engagement = df.groupby("year_month")["total_engagement"].mean().reset_index()
```

```python
print(avg_engagement)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=avg_engagement, x="year_month", y="total_engagement", marker="o")
plt.title("Aylık Ortalama Etkileşim")
plt.xlabel("Ay")
plt.ylabel("Ortalama Etkileşim")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 5️⃣ Platforma Göre Zaman Trendleri
trend = df.groupby(["year_month", "platform"])["n_videos"].sum().reset_index()
```

```python
print(trend)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=trend, x="year_month", y="n_videos", hue="platform", marker="o")
plt.title("Platforma Göre Aylık Video Sayısı")
plt.xlabel("Ay")
plt.ylabel("Video Sayısı")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
